```python
#import dependencies
import pandas as pd
import numpy as np
import datetime as dt

from sqlalchemy import create_engine
import json
from pandas.io.json import json_normalize
from pprint import pprint

#import US data for table
US_youtube_data = "Data/US/USvideos.csv"
CA_youtube_data = "Data/CA/CAvideos.csv"
GB_youtube_data = "Data/GB/GBvideos.csv"

US_youtube = pd.read_csv(US_youtube_data)
CA_youtube = pd.read_csv(CA_youtube_data)
GB_youtube = pd.read_csv(GB_youtube_data)
```

```python
##add in a country code respective to each sheet
US_youtube['country'] = 'United States'
US_youtube.head()
```

```python
CA_youtube['country'] = 'Canada'
CA_youtube.head()
```

```python
GB_youtube['country'] = 'Great Britian'
GB_youtube.head()
```

```python
#merging all of the countries youtube data files
worlds_master = pd.concat([US_youtube, CA_youtube, GB_youtube])
worlds_master.head()
```

```python
#dropping all unnecessary columns for this table
worlds_ytclean = worlds_master.drop(columns=['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed', 'description'])
worlds_ytclean.head()
```

```python
#import US category json file
US_json = "Data/US/US_category_id.json"
CA_json = "Data/CA/CA_category_id.json"
GB_json = "Data/GB/GB_category_id.json"

US_category = pd.read_json(US_json)
CA_category = pd.read_json(CA_json)
GB_category = pd.read_json(GB_json)
```

```python
US_category 

#appending category id to dict
US_cats = []

for x in US_category['items']:
        US_cats.append([x['id'],x['snippet']['title']])
            
US_cats
```

```python
CA_category

#appending category id to dict
CA_cats = []

for x in CA_category['items']:
        CA_cats.append([x['id'],x['snippet']['title']])
            
CA_cats
```

```python
###ran all three in order to ensure that the category sections were the same
GB_category

#appending category id to dict
GB_cats = []

for x in GB_category['items']:
        GB_cats.append([x['id'],x['snippet']['title']])
            
GB_cats
```

```python
#appending the dictionary to a data frame
US_categories = pd.DataFrame.from_dict(US_cats)
US_categories.columns = ['category_id', 'category']
US_categories
```

```python
#reviewing data types to see for necessary conversions
worlds_ytclean.convert_dtypes()
worlds_ytclean.dtypes
```

```python
#reviewing data types to see for necessary conversions
US_categories.convert_dtypes()
US_categories.dtypes
```

```python
#first converting category_id to object type able to be transformed
US_categories['category_id'] = pd.to_numeric(US_categories['category_id'])
```

```python
#merge of both csv and json data
worlds_final = pd.merge(worlds_ytclean, US_categories, on="category_id")
worlds_final.head()
```

```python
#add primary key header for index as to count total youtube video entries, allowing for duplicates
worlds_final.index.name = "youtube_entries"
worlds_final
```

```python
worlds_final['publish_time'] = pd.to_datetime(worlds_final['publish_time'])
worlds_final
```

```python
publish_time = worlds_final['publish_time']
publish_update = pd.to_datetime(publish_time, unit='ms').dt.tz_localize(None)
worlds_final['publish_time'] = publish_update 
```

```python
worlds_final.head()
```

```python
engine = create_engine('postgresql://postgres:0112@localhost:5432/YouTube_Data')
worlds_final.to_sql(name = 'worlds_table', con=engine, if_exists = 'append', index = True)
```
